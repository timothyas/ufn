# Read Replay from Google Cloud Storage (GCS)

```python
import xarray as xr
```

```python
%%time
ds = xr.open_zarr(
    "gcs://noaa-ufs-gefsv13replay/ufs-hr1/1.00-degree/03h-freq/zarr/fv3.zarr",
    storage_options={"token": "anon"},
)
```

```python
%%time
ds.tmp.isel(pfull=-1).sel(time=slice("1995-01-01", "1995-01-02")).plot(col="time", col_wrap=8)
```
